```python
def rentals_month(engine, month, year):
    query= """
SELECT *
FROM rental
WHERE MONTH(rental_date) = %s
AND YEAR(rental_date) = %s
 """

    df_read = pd.read_sql(query, con=engine, params=(month, year))
    return df_read

rental_df= rentals_month(engine, 5, 2005)
```

```python
def rental_count_month(rental_df):
    rental_counts = rental_df.groupby('customer_id').size().reset_index(name='rental_count')
    
    return rental_counts

rental_counts_df = rental_count_month(rental_df)
rental_counts_df
```
